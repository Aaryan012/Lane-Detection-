```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
```

### Converting to grayscale

```python
img_path=cv2.imread(r"road.jpg")
def edge_detect(img_path):
    img=cv2.cvtColor(img_path,cv2.COLOR_BGR2RGB)
    img_gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    # Applying blur
    img_blur=cv2.GaussianBlur(img_gray,(5,5),0)
    # Edge Detection
    img_edge=cv2.Canny(img_blur,150,250)
    return img_edge
```

```python
# Defining Threshold
_,thres=cv2.threshold(img_blur,200,255,cv2.THRESH_BINARY)    
```

```python
plt.imshow(img_edge)
plt.grid()
```

```python
def Return_of_interest(img_edge):
    height,width=img_edge.shape
    mask=np.zeros_like(img_edge)
    polygon=np.array([[
        (0,height),
        (width,height),
        (0,190),
        (width,190),
        (100,165),
        (325,165)
        
    ]],np.int32)
    cv2.fillPoly(mask,polygon,255)
    roi=cv2.bitwise_and(img_edge,mask)
    return roi
```

```python
plt.imshow(Return_of_interest(img_edge))
```

```python
# Using Houghlines to detect lanes 
lines=cv2.HoughLinesP(roi,1,np.pi/180,100,minLineLength=50,maxLineGap=30)
img_hough=img_gray.copy()
left_lines=[] 
right_lines=[] 
if lines is not None:
    for line in lines:
        x1,y1,x2,y2=lines[0]
        slope=(y2-y1)/(x2-x1) if x2 !=x1 else 0
        if slope< -0.5:
            left_lines.append(line[0])
        elif slope > 0.5:
            right_lines.append(line[0])
plt.imshow(img_hough)
```

```python
contours,_=cv2.findContours(thres,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
img_cont=img_gray.copy()
for cnt in contours:
    if cv2.contourArea(cnt)>1000:
        cv2.drawContours(img_cont,[cnt],-1,(0,255,0),2)
plt.imshow(img_cont,cmap='gray')
```
